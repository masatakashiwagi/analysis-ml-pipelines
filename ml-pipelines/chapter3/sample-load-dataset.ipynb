```python
import os
import pandas as pd
from tfx.components import CsvExampleGen
from tfx.orchestration.experimental.interactive.interactive_context import InteractiveContext
```

## ローカルストレージからデータセットを読み込む
### CSVファイルをtf.train.Example型に変換
- `data`ディレクトリにあるcsvファイルをロードする
- ※事前にダウンロードスクリプトを実行して，csvファイルをダウンロードしておく

```python
base_dir = os.getcwd()
data_dir = os.path.join(base_dir, os.pardir, 'data/csv')
example_gen = CsvExampleGen(input_base=data_dir)

context = InteractiveContext()
context.run(example_gen)
```

### TFRecordファイルをインポートする場合
- TFRecordファイルは既にtf.train.Example型に変換されているので，変換なしで取り込める

```python
from tfx.components import ImportExampleGen
```

```python
base_dir = os.getcwd()
data_dir = os.path.join(base_dir, os.pardir, 'tfrecord_data')
example_gen = ImportExampleGen(input_base=data_dir)

context.run(example_gen)
```

### Parquet/Avroをtf.train.Example型に変換する場合
- 汎用的系なファイル読み込みコンポーネントであるFileBasedExampleGenを使う
- executer_classをオーバーライドして使う

```python
from tfx.components import FileBasedExampleGen
from tfx.components.example_gen.custom_executors import parquet_executor, avro_executor
```

```python
# Parquet
base_dir = os.getcwd()
data_dir = os.path.join(base_dir, os.pardir, 'parquet_data')
example_gen = FileBasedExampleGen(
    input_base=data_dir,
    executer_class=parquet_executor
)

context.run(example_gen)
```

```python
# Avro
base_dir = os.getcwd()
data_dir = os.path.join(base_dir, os.pardir, 'avro_data')
example_gen = FileBasedExampleGen(
    input_base=data_dir,
    executer_class=avro_executor
)

context.run(example_gen)
```
